```python
## Importing SEABORN
import seaborn as sns
```

```python
import matplotlib as plt
```

```python
import numpy as np 
import pandas as pd
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
tips.describe()
```

```python
tips.head()
```

```python
#Visualise the data
sns.set_theme()
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    col='time',        # Creates separate plots for lunch and dinner
    hue='smoker',      # Different colors for smokers and non-smokers
    style='smoker',    # Different markers for smokers and non-smokers
    size='size',       # Adjusts the size of the markers based on the size of the party
    kind='scatter'     # Ensure it's a scatter plot
)
plt.show()
```
