```python
x=np.linspace(0,10,100)
```

```python
x
```

```python
y=np.sin(x)
```

```python
z=np.cos(x)
```

```python
## Plotting the data
mp.plot(x,y)
mp.show()
```

```python
mp.plot(x,z)
mp.xlabel=('angle')
mp.ylabel=('sine')
mp.title=('Sine Wave')
mp.show()
```

```python
##Bar Chart

fig=mp.figure()
axes=fig.add_axes([0,0,1,1])
lang=['Englissh','Chinese','Bengali','Hindi','Marathi']
people=[100,200,43,544,23]
axes.bar(lang,people)
mp.xlabel=('Language')
mp.ylabel=('No of people')
mp.show()
```

```python
##Pie Chart
figure=mp.figure()
axes=figure.add_axes([0,0,1,1])
lang=['Englissh','Chinese','Bengali','Hindi','Marathi']
people=[100,200,43,544,23]
axes.pie(people,labels=lang,autopct='%1.1f')
mp.show()
```

```python
##Scatterplot
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
gi=mp.figure()
axes=gi.add_axes([0,0,1,1])
axes.scatter(x,y,c='green')
axes.scatter(x,z,c='b')
mp.show()
```

```python
##3D Scatter PLot
figure=mp.figure()
axes=mp.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
axes.scatter(x,y,z,c=z,cmap='turbo_r')
```
